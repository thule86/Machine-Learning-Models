```python
import pandas as pd
import math
import numpy as np
from scipy import stats
np.random.seed(200)
a = np.random.normal(0, 1, 100)
b = np.random.normal(1.5, math.sqrt(1.1), 100)
x = np.concatenate((a, b))
df1 = pd.DataFrame(x, columns = ['x'])
df1['x'].plot(title = 'Observation')
```

```python
## Plot log likelihood Ratio
df1['log1'] = (math.log(1/math.sqrt(1.1))) + ((1/2-1/2.2) * (df1['x']**2)) + ((3/2.2)*df1['x']) - (2.25/2.2)
import matplotlib.pyplot as plt
df1['Wt_1'] = 0
df1['log1'].plot(title = 'Log Likelihood Ratio')
```

```python
## Plot CUSUM
df1['Wt'] = df1['log1'].cumsum()
df1["Wt2"] = df1[['Wt', 'Wt_1']].values.max(1)
df1['Wt2'].plot(title = 'CUSUM Statistic')
```
