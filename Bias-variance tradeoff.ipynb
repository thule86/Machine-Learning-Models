```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 
```

```python
X = np.array([[1,0.7], [1,1]])
x_test1 =np.array([[1, 0.5]])
x_test2 =np.array([[1, 1.5]])
beta = np.array([1,1])
W=np.diag([1/1.5,1])
I = np.eye((X.shape[0]))
```

```python
def plot_graph(x_test):
    alphas = np.arange(0.1,3,0.1)
    bias_1 =[]
    var_1 =[]
    mse_1 = []
    for i in alphas:
        U = np.linalg.inv(X.T@W@X + i * I)
        b1 =np.square(-i*x_test@U@beta)
        bias_1.append([i,b1])  
        U2 = U@X.T@W
        v1 = x_test@U2@np.linalg.inv(W)@U2.T@x_test.T
        var_1.append([i,v1])  
        z1 = b1 + v1 
        mse_1.append([i,z1]) 
    print('optimal alpha and mse is',min(mse_1, key = lambda t: t[1]))    
    plt.plot(np.array(mse_1)[:,0], np.array(mse_1)[:,1],color="r", label='MSE')
    plt.plot(np.array(var_1)[:,0], np.array(var_1)[:,1],color="g", label='Variance')
    plt.plot(np.array(bias_1)[:,0], np.array(bias_1)[:,1],color="b", label='Bias-squared')
    plt.legend(loc="upper right")
    plt.xlabel("alpha")
    plt.ylabel("MSE")
    plt.title("x ="+ str(x_test)+":  Bias-squared, Variance, MSE vs Alpha")
    plt.show() 
    
```

```python
plot_graph(x_test1)
```

```python
plot_graph(x_test2)
```
